```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the TSV file
file_path = 'results/gpt-3.5-turbo/end_to_end.tsv'
data = pd.read_csv(file_path, sep='\t')
```

```python
def plot(data, col, title, savepath):
# Set a style for the plot
    plt.style.use('seaborn-darkgrid')

    # Define the custom bins you want from -4 to 4
    bins = [-4.5, -3.5, -2.5, -1.5, -0.5, 0.5, 1.5, 2.5, 3.5, 4.5]

    # Create a figure and axis
    plt.figure(figsize=(10, 6))

    # Customize the histogram with your specified bins
    sns.histplot(data[col].dropna(), bins=bins, kde=False)

    # Add titles and labels with improved font sizes and descriptions
    plt.title(title, fontsize=28, fontweight='bold')
    plt.xlabel('Moral Score', fontsize=28)
    plt.ylabel('Frequency', fontsize=28)

    # Customize ticks
    plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4], fontsize=20)
    plt.yticks(fontsize=20)

    # Add a grid with custom alpha and linestyle for visibility
    plt.grid(True, alpha=0.3, linestyle='--')

    # Display the plot
    # plt.show()
    if savepath:
        plt.savefig(savepath, dpi="figure")
```

```python
plot(data, "moral_score", "Distribution of vanilla model predictions", "plots/end_to_end.png")

dyadic_data = pd.read_csv('results/gpt-3.5-turbo/dyadic.tsv', sep='\t')
```

```python
plot(data, "human_score", "Distribution of Human scores", "plots/human.png")
```

```python
plot(dyadic_data, "moral_score", "Distribution of  Dyadic Bottleneck model predictions", "plots/dyadic_bottleneck.png")
```
